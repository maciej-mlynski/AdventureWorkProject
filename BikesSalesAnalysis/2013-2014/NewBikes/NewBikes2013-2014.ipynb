```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
sns.set(rc={'figure.figsize':(10,6)})
```

```python
df = pd.read_csv('./../GroupedData/2013-2014_Gr_Unavlb_2012_sub_class_color_size.csv', sep=';')
```

```python
df['totalSellAmt'] = df[['totalSellAmt']].applymap(lambda x: str(x.replace(',','.'))).astype('float')
df['totalCosts'] = df[['totalCosts']].applymap(lambda x: str(x.replace(',','.'))).astype('float')
df['totalGrossProfit'] = df[['totalGrossProfit']].applymap(lambda x: str(x.replace(',','.'))).astype('float')
```

```python
bikes = df[df['Size'].isnull() & df['Color'].isnull() & df['Class'].isnull()]
```

```python
plt.pie(bikes['totalQuantity'], autopct = '%0.0f%%', labels =bikes['SubName']);
```

```python
sns.barplot(x=bikes['SubName'], y=bikes['totalGrossProfit'])
```

```python
sns.barplot(x=bikes['SubName'], y=bikes['totalQuantity'])
```

As we can see Touring bikes has been sold the most often, but doesn't translate into results, becouse total gross profit is the lowest for 3 groups. Suprising is that the Road Bikes brings the highest income for company. Obviosly, we consider here only new bikes, but if we took all the Mountain bikes will be first. Let's dive deaper

```python
name_class = df[df['Size'].isnull() & df['Color'].isnull() & df['Class'].notnull()]
```

```python
name_class['SubName'].iloc[0] = 'Mountain Bikes_L'
name_class['SubName'].iloc[1] = 'Mountain Bikes_M'

name_class['SubName'].iloc[2] = 'Road Bikes_L'
name_class['SubName'].iloc[3] = 'Road Bikes_M'

name_class['SubName'].iloc[4] = 'Touring Bikes_H'
name_class['SubName'].iloc[5] = 'Touring Bikes_L'
name_class['SubName'].iloc[6] = 'Touring Bikes_M'
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
sns.barplot(x=name_class['SubName'], y=name_class['totalGrossProfit'])
y_value=['{:,.0f}'.format(x).replace(",", " ") + 'PLN' for x in ax.get_yticks()]
ax.set_yticklabels(y_value)
plt.xticks(rotation=30);
```

As we can see only one class of bikes has negative income, which is Touring Bike class L - LOW. Fortunetly for this model High class model has the highest income that covers the losses

```python
df = pd.read_csv('./../GroupedData/2013-2014_Gr_Unavlb_2012_model_color_size.csv', sep=';')
```

```python
df['totalSellAmt'] = df[['totalSellAmt']].applymap(lambda x: str(x.replace(',','.'))).astype('float')
df['totalCosts'] = df[['totalCosts']].applymap(lambda x: str(x.replace(',','.'))).astype('float')
df['totalGrossProfite'] = df[['totalGrossProfite']].applymap(lambda x: str(x.replace(',','.'))).astype('float')
```

```python
Models = df[df['Size'].isnull() & df['Color'].isnull()]
```

```python
Models
```

Company delivered 7 different models of bikes in 2013

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
sns.barplot(x=Models['ModelName'], y=Models['totalGrossProfite'])
y_value=['{:,.0f}'.format(x).replace(",", " ") + ' USD' for x in ax.get_yticks()]
ax.set_yticklabels(y_value)
plt.xticks(rotation=30);
```

```python
model_color = df[df['Color'].notnull() & df['Size'].isnull() & df['ModelName'].notnull()]
```

```python
model_color
```

```python
model_color['ModelName_color'] = model_color[['ModelName', 'Color']].agg('-'.join, axis=1)
```

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
sns.barplot(x=model_color['ModelName_color'], y=model_color['totalGrossProfite'])
y_value=['{:,.0f}'.format(x).replace(",", " ") + 'PLN' for x in ax.get_yticks()]
ax.set_yticklabels(y_value)
plt.xticks(rotation=60);
```
