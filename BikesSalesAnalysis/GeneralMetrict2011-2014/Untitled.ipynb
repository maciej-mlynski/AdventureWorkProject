```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
pd.set_option('display.max_columns', None)

sns.set(rc={'figure.figsize':(12,6)})
```

```python
df1 = pd.read_csv('./../../PredictingYearlyOrderQuantity/Data/2011-2012OrdersQuantity.csv', sep=';')
df2 = pd.read_csv('./../../PredictingYearlyOrderQuantity/Data/2012-2013OrdersQuantity.csv', sep=';')
df3 = pd.read_csv('./../../PredictingYearlyOrderQuantity/Data/2013-2014OrdersQuantity.csv', sep=';')
```

```python
df = pd.concat([df1,df2,df3])
```

```python
df['UnitPriceDiscount'] = df[['UnitPriceDiscount']].applymap(lambda x: str(x.replace(',','.'))).astype('float64')
df['StanstadCost'] = df[['StanstadCost']].applymap(lambda x: str(x.replace(',','.'))).astype('float64')
df['ListPrice'] = df[['ListPrice']].applymap(lambda x: str(x.replace(',','.'))).astype('float64')
df['FinalSellPrice'] = df[['FinalSellPrice']].applymap(lambda x: str(x.replace(',','.'))).astype('float64')
df['PriceToCost'] = df[['PriceToCost']].applymap(lambda x: str(x.replace(',','.'))).astype('float64')
```
