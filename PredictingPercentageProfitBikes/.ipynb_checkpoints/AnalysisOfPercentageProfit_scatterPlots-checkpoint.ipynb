```python
# Representant values:
# Color: Black
# Class: High
# Order year: 2011Orders
# SubName: MountainBikes
# YearOfAvelliability: LessThanYAveliable
```

```python
X = df[['Size', 'Weight', 'StandardCost',
       'isDiscounted', 
        'Silver', 'Red', 'Yellow', 'Blue', 'L', 'M',
       'OneYAbvl', 'TwoYaveliable',
       '2012Orders', '2013Orders', '2014Orders',
       'Road Bikes', 'Touring Bikes' 
]]
```

```python
y = df['PercentProfit']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train = scaler.transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train,y_train)
```

```python
test_predictions = model.predict(X_test)
```

```python
test_predictions
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
MAE = mean_absolute_error(y_test,test_predictions)
MSE = mean_squared_error(y_test,test_predictions)
RMSE = np.sqrt(MSE)

print(f'MAE : {MAE} \t MSE : {MSE} \t RMSE : {RMSE}')
```
