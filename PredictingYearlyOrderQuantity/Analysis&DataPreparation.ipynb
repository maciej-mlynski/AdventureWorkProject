I have decided that predicting total profit from singular bike could be unusefull mainly becouse changing the unit price leads to increase or decrease the quantity of orders. To put it simply we can say that Adventure shop cut the prices to sell more bikes and it do not work in oposite way. If bicycle demand grow company should even increase the price. **So, my conclusion is that we should inlude different unit prices for each bike and try to predict total quantity of sale, base on past experiance.**

#### Dummie variables

```python
df['Black'] = np.where((df['Color'] == 'Black'), 1, 0)
df['Silver'] = np.where((df['Color'] == 'Silver'), 1, 0)
df['Red'] = np.where((df['Color'] == 'Red'), 1, 0)
df['Yellow'] = np.where((df['Color'] == 'Yellow'), 1, 0)
df['Blue'] = np.where((df['Color'] == 'Blue'), 1, 0)
```

```python
df['H'] = np.where((df['Class'] == 'H '), 1, 0)
df['L'] = np.where((df['Class'] == 'L '), 1, 0)
df['M'] = np.where((df['Class'] == 'M '), 1, 0)
```

```python
df['YearsOfAveliability'].unique()
```

```python
df['isNew'] = df['YearsOfAveliability']
```

```python
df['2011Sales'] = np.where((df['YearsOfAnalyze'] == '06.2011-06.2012'), 1, 0)
df['2012Sales'] = np.where((df['YearsOfAnalyze'] == '06.2012-06.2013'), 1, 0)
df['2013Sales'] = np.where((df['YearsOfAnalyze'] == '06.2013-06.2014'), 1, 0)
```

```python
df['Mountain Bikes'] = np.where((df['Name'] == 'Mountain Bikes'), 1, 0)
df['Road Bikes'] = np.where((df['Name'] == 'Road Bikes'), 1, 0)
df['Touring Bikes'] = np.where((df['Name'] == 'Touring Bikes'), 1, 0)
```

```python
df['Jan'] = np.where((df['MonthOfSale'] == 1), 1, 0)
df['Feb'] = np.where((df['MonthOfSale'] == 2), 1, 0)
df['Mar'] = np.where((df['MonthOfSale'] == 3), 1, 0)
df['Apr'] = np.where((df['MonthOfSale'] == 4), 1, 0)
df['May'] = np.where((df['MonthOfSale'] == 5), 1, 0)
df['Jun'] = np.where((df['MonthOfSale'] == 6), 1, 0)
df['Jul'] = np.where((df['MonthOfSale'] == 7), 1, 0)
df['Aug'] = np.where((df['MonthOfSale'] == 8), 1, 0)
df['Sep'] = np.where((df['MonthOfSale'] == 9), 1, 0)
df['Oct'] = np.where((df['MonthOfSale'] == 10), 1, 0)
df['Nov'] = np.where((df['MonthOfSale'] == 11), 1, 0)
df['Dec'] = np.where((df['MonthOfSale'] == 12), 1, 0)
```

```python
df['Isdiscounted'] = np.where((df['UnitPriceDiscount'] > 0.02), 1, 0)
```

```python
df['IsPriceHigherThanCosts'] = np.where((df['PriceToCost'] >= 1), 1, 0)
```

```python
df.columns
```

```python
df = df[['Size', 
       'YearsOfAveliability', 'Weight', 'StanstadCost',
       'ListPrice', 'FinalSellPrice', 'PriceToCost', 'OrdersQuantity',
       'Black', 'Silver', 'Red', 'Yellow', 'Blue', 'H', 'L',
       'M', 'isNew', '2011Sales', '2012Sales', '2013Sales',
       'Mountain Bikes', 'Road Bikes', 'Touring Bikes', 'Jan', 'Feb', 'Mar',
       'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec',
       'Isdiscounted', 'IsPriceHigherThanCosts']].copy()
```
